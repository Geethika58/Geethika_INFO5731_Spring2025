# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
import math
def hemisphere_volume(r):
    return (2/3) * math.pi * (r ** 3)
print("r\tV")
for r in range(2, 21, 2):
    volume = hemisphere_volume(r)
    print(f"{r}\t{volume:.2f}")
```

# Question 3 (10 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
# Your code here
import string
# (a) Prompt the user to input two sentences
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# (b) Transform each sentence into a list of words
def sentence_to_word_list(sentence):
    # Using str.split() to break the sentence into words, preserving punctuation
    words = sentence.split()
    return words
words1 = sentence_to_word_list(sentence1)
words2 = sentence_to_word_list(sentence2)

print("\nList of words from Sentence 1:")
print(words1)
print("\nList of words from Sentence 2:")
print(words2)

# (c) Compare the two sentences and find common words
common_words = list(set(words1) & set(words2))
print("\nCommon words in both sentences:")
print(common_words)

# (d) Remove punctuation from the sentences
def remove_punctuation(sentence):
    # Remove punctuation using string.punctuation which contains common punctuation marks
    translator = str.maketrans('', '', string.punctuation)
    return sentence.translate(translator)
clean_sentence1 = remove_punctuation(sentence1)
clean_sentence2 = remove_punctuation(sentence2)
print("\nSentence 1 without punctuation:")
print(clean_sentence1)
print("\nSentence 2 without punctuation:")
print(clean_sentence2)

# (e) Count the number of vowels in each sentence
def count_vowels(sentence):
    vowels = "aeiou"
    vowel_count = {v: sentence.lower().count(v) for v in vowels}
    return vowel_count
vowels_in_sentence1 = count_vowels(clean_sentence1)
vowels_in_sentence2 = count_vowels(clean_sentence2)

print("\nVowel counts in Sentence 1:")
print(vowels_in_sentence1)
print("\nVowel counts in Sentence 2:")
print(vowels_in_sentence2)
```

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
# Your code here
import math
# Caluclator logic function
def calculator():
    print("Welcome to the Python Command-Line Calculator!")
    print("Supported operations: +, -, *, /, ** (exponentiation), sqrt (square root), % (modulus), parentheses ()")
    print("Type 'exit' to quit the calculator.")
    while True:
        # Prompt the user for input and evaluate the expression
        expression = input("\nEnter a mathematical expression: ")
        # Exit condition for the calculator
        if expression.lower() == 'exit':
            print("You have exited the calculator.")
            break
        expression = expression.replace("^", "**")
        # Handle errors and evaluate the expression
        try:
            expression = expression.replace("sqrt", "math.sqrt")
            result = eval(expression)
            # Display the result
            print(f"Result: {result}")
        except ZeroDivisionError:
            print("Error: Division by zero is not allowed.")
        except (SyntaxError, NameError):
            print("Error: Invalid expression. Please try again.")
        except Exception as e:
            print(f"Unexpected error: {e}")
# Run the calculator
calculator()
```

# Question 5 (20 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
 # Your code here
import csv
# Read the CSV data and extract necessary fields
housing_data = {}
vacant_units = {}
occupied_units = {}
total_units = {}
with open('Assignment1_denton_housing.csv', mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        year = int(row['year'])
        title = row['title_field']
        value = int(row['value'])
        # Store the values for vacant, occupied, and total units
        if title == "Occupied Housing Units":
            housing_data[year] = value
            occupied_units[year] = value
        elif title == "Vacant Housing Units":
            vacant_units[year] = value
        elif title == "Total Housing Units":
            total_units[year] = value

# (a) Calculate the Difference in Occupied Housing Units Year to Year
differences = []
years = sorted(housing_data.keys())
for i in range(1, len(years)):
    year_diff = years[i] - years[i - 1]
    diff_value = housing_data[years[i]] - housing_data[years[i - 1]]
    differences.append((year_diff, diff_value))
# Sort the differences by the value
differences.sort(key=lambda x: x[1])
print("Differences in Occupied Housing Units year by year (sorted):")
for year_diff, diff_value in differences:
    print(f"{years[year_diff-1]} - {years[year_diff]}: {diff_value}")

# (b) Calculate the Percentage of Vacant and Occupied Housing Units
print("\nYear Vacant Housing Units Occupied Housing Units")
for year in sorted(total_units.keys()):
    vacant_percent = (vacant_units[year] / total_units[year]) * 100
    occupied_percent = (occupied_units[year] / total_units[year]) * 100
    print(f"{year} {vacant_percent:.2f}% {occupied_percent:.2f}%")

# (c) Find the Year and Value of the Highest Number of Vacant and Occupied Housing Units
max_vacant_year = max(vacant_units, key=vacant_units.get)
max_occupied_year = max(occupied_units, key=occupied_units.get)
print("\nYear and Value of Maximum Vacant and Occupied Housing Units:")
print(f"Vacant Housing Units: {max_vacant_year} {vacant_units[max_vacant_year]}")
print(f"Occupied Housing Units: {max_occupied_year} {occupied_units[max_occupied_year]}")

# (d) Calculate the Harmonic Mean of Total Housing Units
total_values = list(total_units.values())
n = len(total_values)
harmonic_mean = n / sum(1 / x for x in total_values)
print(f"\nHarmonic Mean of Total Housing Units: {harmonic_mean:.2f}")
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
# Your code here
"""
Prompt that I used:  I need help to complete my assignment to Develop a advance Python program to manage inventory records for a prodcut managing company.
 The system needs functions for adding, deleting and changing products and basic search options.
 Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files.
 Finally maintaine 50 products of different categories in inventory."""

import csv
import os

# File path for the inventory CSV
inventory_file = 'inventory.csv'

# Default categories
categories = ['Electronics', 'Clothing', 'Furniture', 'Groceries', 'Books']

# Function to read the inventory from CSV
def read_inventory():
    inventory = []
    if os.path.exists(inventory_file):
        with open(inventory_file, mode='r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                inventory.append(row)
    return inventory

# Function to write inventory to CSV
def write_inventory(inventory):
    with open(inventory_file, mode='w', newline='') as file:
        fieldnames = ['product_name', 'product_id', 'category', 'price', 'stock']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(inventory)

# Function to add a product to the inventory
def add_product():
    inventory = read_inventory()

    product_name = input("Enter product name: ")
    product_id = input("Enter product ID (unique): ")

    # Ensure product_id is unique
    if any(product['product_id'] == product_id for product in inventory):
        print("Error: Product ID already exists.")
        return

    category = input(f"Enter product category from {categories}: ")
    if category not in categories:
        print("Error: Invalid category. Choose a valid category.")
        return

    try:
        price = float(input("Enter product price: "))
        stock = int(input("Enter product stock quantity: "))
    except ValueError:
        print("Error: Invalid price or stock input.")
        return

    new_product = {
        'product_name': product_name,
        'product_id': product_id,
        'category': category,
        'price': price,
        'stock': stock
    }

    inventory.append(new_product)
    write_inventory(inventory)
    print(f"Product {product_name} added successfully!")

# Function to delete a product
def delete_product():
    inventory = read_inventory()

    product_id = input("Enter product ID to delete: ")

    # Find product by product_id
    product_to_delete = next((product for product in inventory if product['product_id'] == product_id), None)

    if not product_to_delete:
        print("Error: Product not found.")
        return

    inventory.remove(product_to_delete)
    write_inventory(inventory)
    print(f"Product {product_to_delete['product_name']} deleted successfully!")

# Function to update a product's details
def update_product():
    inventory = read_inventory()

    product_id = input("Enter product ID to update: ")

    # Find the product
    product_to_update = next((product for product in inventory if product['product_id'] == product_id), None)

    if not product_to_update:
        print("Error: Product not found.")
        return

    print(f"Current details: {product_to_update}")

    new_name = input("Enter new name (leave blank to keep current): ")
    new_category = input(f"Enter new category from {categories} (leave blank to keep current): ")

    if new_name:
        product_to_update['product_name'] = new_name
    if new_category and new_category in categories:
        product_to_update['category'] = new_category

    try:
        new_price = input("Enter new price (leave blank to keep current): ")
        if new_price:
            product_to_update['price'] = float(new_price)

        new_stock = input("Enter new stock (leave blank to keep current): ")
        if new_stock:
            product_to_update['stock'] = int(new_stock)
    except ValueError:
        print("Error: Invalid price or stock input.")
        return

    write_inventory(inventory)
    print(f"Product {product_to_update['product_name']} updated successfully!")

# Function to search for a product
def search_product():
    inventory = read_inventory()

    search_term = input("Enter product name or category to search: ").lower()

    results = [product for product in inventory if search_term in product['product_name'].lower() or search_term in product['category'].lower()]

    if not results:
        print("No products found.")
    else:
        print(f"Found {len(results)} product(s):")
        for result in results:
            print(result)

# Function to restock products
def restock_product():
    inventory = read_inventory()

    product_id = input("Enter product ID to restock: ")

    product_to_restock = next((product for product in inventory if product['product_id'] == product_id), None)

    if not product_to_restock:
        print("Error: Product not found.")
        return

    try:
        restock_qty = int(input(f"Enter quantity to add to stock for {product_to_restock['product_name']}: "))
        if restock_qty < 0:
            print("Error: Cannot restock with negative quantity.")
            return
        product_to_restock['stock'] = int(product_to_restock['stock']) + restock_qty  # Ensure stock is an integer
    except ValueError:
        print("Error: Invalid quantity input.")
        return

    write_inventory(inventory)
    print(f"Product {product_to_restock['product_name']} restocked successfully!")

# Function to generate a report
def generate_report():
    inventory = read_inventory()

    print("Inventory Report:")
    category_totals = {}
    total_value = 0

    for product in inventory:
        # Ensure price is a float and stock is an integer
        price = float(product['price'])
        stock = int(product['stock'])

        category_totals[product['category']] = category_totals.get(product['category'], 0) + price * stock
        total_value += price * stock

    for category, value in category_totals.items():
        print(f"{category}: ${value:.2f}")

    print(f"Total inventory value: ${total_value:.2f}")

# Edge case: Handle corrupted CSV (invalid or missing fields)
def handle_corrupted_csv():
    try:
        read_inventory()
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        print("Attempting to create a new inventory file...")
        write_inventory([])  # Create a new file if corrupted

# Main menu function
def main():
    handle_corrupted_csv()  # Ensure no corrupted files exist

    while True:
        print("\nWelcome to Inventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. Restock Product")
        print("6. Generate Report")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_product()
        elif choice == '2':
            delete_product()
        elif choice == '3':
            update_product()
        elif choice == '4':
            search_product()
        elif choice == '5':
            restock_product()
        elif choice == '6':
            generate_report()
        elif choice == '7':
            break
        else:
            print("Invalid choice. Please try again.")

# Run the main program
main()
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
# Your code here
"""
Prompt that I used: Create a well designed Python program for a flight booking system.
Offer core functions for adding new flights, finding trips, reserving and
releasing seating space and creating performance reports.
Ensure the system manages edge cases like invalid input, overbooking, and cancellations.
Finally, extend the booking system by including advanced features and maintain about 50 flight in the system."""
import csv
import os
# File path for the flights CSV
flight_file = 'flights.csv'

# Function to read flight data from the CSV file
def read_flights():
    flights = []
    if os.path.exists(flight_file):
        with open(flight_file, mode='r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                flights.append(row)
    return flights

# Function to write flight data to the CSV file
def write_flights(flights):
    with open(flight_file, mode='w', newline='') as file:
        fieldnames = ['flight_number', 'destination', 'total_seats', 'available_seats', 'price']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(flights)

# Function to add a new flight to the system
def add_flight():
    flights = read_flights()
    flight_number = input("Enter flight number (unique): ")

    # Ensure the flight number is unique
    if any(flight['flight_number'] == flight_number for flight in flights):
        print("Error: Flight number already exists.")
        return

    destination = input("Enter flight destination: ")
    try:
        total_seats = int(input("Enter total number of seats: "))
        available_seats = total_seats  # initially, all seats are available
        price = float(input("Enter flight price: "))
    except ValueError:
        print("Error: Invalid number for seats or price.")
        return

    new_flight = {
        'flight_number': flight_number,
        'destination': destination,
        'total_seats': total_seats,
        'available_seats': available_seats,
        'price': price
    }

    flights.append(new_flight)
    write_flights(flights)
    print(f"Flight {flight_number} added successfully!")

# Function to find a flight based on destination or flight number
def find_flight():
    flights = read_flights()
    search_term = input("Enter flight number or destination to search: ").lower()

    results = [flight for flight in flights if search_term in flight['flight_number'].lower() or search_term in flight['destination'].lower()]

    if not results:
        print("No flights found.")
    else:
        print(f"Found {len(results)} flight(s):")
        for result in results:
            print(result)

# Function to reserve seats for a flight
def reserve_seat():
    flights = read_flights()
    flight_number = input("Enter flight number to reserve seats: ")

    flight = next((f for f in flights if f['flight_number'] == flight_number), None)

    if not flight:
        print("Error: Flight not found.")
        return

    try:
        seats_to_reserve = int(input(f"Enter number of seats to reserve (Available: {flight['available_seats']}): "))
        if seats_to_reserve <= 0:
            print("Error: You must reserve at least one seat.")
            return
        if seats_to_reserve > int(flight['available_seats']):
            print("Error: Not enough available seats.")
            return
    except ValueError:
        print("Error: Invalid seat number.")
        return

    flight['available_seats'] = str(int(flight['available_seats']) - seats_to_reserve)
    write_flights(flights)
    print(f"{seats_to_reserve} seat(s) reserved successfully for flight {flight_number}.")

# Function to release reserved seats (cancellation)
def cancel_reservation():
    flights = read_flights()
    flight_number = input("Enter flight number to cancel reservation: ")

    flight = next((f for f in flights if f['flight_number'] == flight_number), None)

    if not flight:
        print("Error: Flight not found.")
        return

    try:
        seats_to_release = int(input(f"Enter number of seats to cancel (Reserved: {int(flight['total_seats']) - int(flight['available_seats'])}): "))
        if seats_to_release <= 0:
            print("Error: You must cancel at least one seat.")
            return
        if seats_to_release > int(flight['total_seats']) - int(flight['available_seats']):
            print("Error: Not enough seats to cancel.")
            return
    except ValueError:
        print("Error: Invalid seat number.")
        return

    flight['available_seats'] = str(int(flight['available_seats']) + seats_to_release)
    write_flights(flights)
    print(f"{seats_to_release} seat(s) canceled successfully for flight {flight_number}.")

# Function to generate performance report for the flights
def generate_report():
    flights = read_flights()

    total_flights = len(flights)
    total_seats = sum(int(flight['total_seats']) for flight in flights)
    total_available_seats = sum(int(flight['available_seats']) for flight in flights)
    total_booked_seats = total_seats - total_available_seats

    print("\nFlight Performance Report:")
    print(f"Total flights: {total_flights}")
    print(f"Total seats available: {total_available_seats}")
    print(f"Total seats booked: {total_booked_seats}")
    print(f"Total seats in all flights: {total_seats}")
    print(f"Total booking percentage: {total_booked_seats / total_seats * 100:.2f}%")

# Main menu function to run the system
def main():
    while True:
        print("\nFlight Booking System")
        print("1. Add Flight")
        print("2. Find Flight")
        print("3. Reserve Seats")
        print("4. Cancel Reservation")
        print("5. Generate Report")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_flight()
        elif choice == '2':
            find_flight()
        elif choice == '3':
            reserve_seat()
        elif choice == '4':
            cancel_reservation()
        elif choice == '5':
            generate_report()
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please try again.")

# Run the main program
main()
```
